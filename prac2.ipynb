```python
# Normalisation for "exam_score" column

df["exam_score_n"]=(df["exam_score"]-df["exam_score"].mean())/df["exam_score"].std()
```

```python
# to fill the null values with mean
df["age"].fillna(df["age"].mean())
```

```python
# to fill the null values with mode 
df["sleep_hours"].fillna(df["sleep_hours"].mode())
```

```python
# to find IQR
Q1 = df['class_attendance'].quantile(0.25)
Q3 = df['class_attendance'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```
